```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
df_scaled
```

```python
selector = SelectKBest(score_func=f_classif, k=10)
X_new = selector.fit_transform(df_scaled.drop('target', axis=1), df_scaled['target'])
selected_features_anova = df.columns[:-1][selector.get_support()]
print("Selected feature with ANOVA:", selected_features_anova)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, df_scaled['target'], test_size=0.2, random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Save the model
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
```
